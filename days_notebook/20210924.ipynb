# 주식회사 역사와 의미

- `유한책임`과 `지분양도`를 보장함으로써, 일반 대중으로부터 영세 자금을 모아서 대규모 자금을 만드는데 성공할 수 있었다.

- 단순한 제로섬 게임이 아니다.

- 대중에게 판매된 최초의 주식은 1602년 네덜란드 동인도 회사 주식이다.

- 주식회사의 의미
    + 영어표기 : Company(Co.), Corporation(Corp.), Incorporated(Inc.), Limited(Ltd.) 등
    + Corporation, Incorporated 는 몸을 뜻하는 라틴어인 corpus에서 파생된 단어로, 법인으로 등록되어 법률상 권리와 의무를 가지는 주체임을 의미한다.
    + Limited 는 책임이 유한하다는 뜻인데, 회사가 잘못되거나 큰 부채가 발생하도 주주의 책임은 보유한 주식의 범위로 한정됨을 말한다.



# 우리나라의 증권시장

- 증시역사 : 1896년 인천미두취인소(선물 거래소), 약 120여년 정도 됨, 취인소 = 거래소

- 최초 증권 거래시장 : 1920년 경성주식현물취인소

- 1932년 조선취인소(인천미두취인소 합병)

- 1956년 대한증권거래소 -> 한국거래소(Korean exchange, KRX)

- 1983년 한국종합주가지수(KOSPI) 발표, 1980년 1월 4일에 상장된 모든 종목의 시가총액을 기준 지수인 100포인트로 집계했다.
    + KOSPI 지수 2500 = 한국 증시 총액이 1980년 당시보다 25배
    + KOSPI 하락폭이 컸던 시점 : 1997년 IMF 외횐위기 / 2008년 서브프라임 금융위기


# 워렌 버핏과 가치 투자

- 워렌 버핏의 주식투자 규칙
    + 규착 1. 절대 돈을 잃지마라.
    + 규착 2. 규칙 1을 절대 잊지마라.
    
- 투자와 투기
    + > 투자란 안전하게 원금을 보장하면서도 적절한 수익을 돌려줄 수 있는지 분석하는 과정에서 이루어진다. 이러한 조건을 만족하지 못하는 것들은 투기다.
    
- 버크셔 해서웨이의 주주 서한
    + 워렌 버핏이 매년 작성한다. 그의 사상을 가장 정확히 이해하려면 해당 글을 읽어볼것!
    + 벤저민 그레이엄 : 정량적 분석, 시가총액이 순당좌자산의 2/3이하로 저평가된 종목에 분산 투자
    + 워렌 버핏 : 경영자의 태도나 브랜드 충성도 같이 계량화하기 어려운 정성적 요소도 고려하여 투자
        + > 그저 그런 회사를 매우 싸게 사는 것보다 유명한 회사를 적당한 가격에 사는 것이 훨씬 낫다.
        
- 주식과 채권의 포트폴리오 전략
    + 원금보장 & 적절한 수익을 돌려주는 투자 기법? -> 증액투자법 : 채권에 투자해서 얻은 이자 수익으로 주식에 재투자하는 기법
    + 채권수익률은 주식보다는 떨어짐. 발행주체가 망하지 않는 한 수익이 보장된다.
    + 이외에도 정률투자법 이라는 것도 존재함.

# 얼마나 벌 것인가?

- 장기 투자와 수익률
    + 한국투자밸류자산운용 이채원 부사장 : "KOSPI 지수가 낮을 때 투자하는 것도 중요하지만, 지수가 높을 때 투자하더라도 장기간 투자하는 게 성과가 훨씬 좋았다."
    + 수익이 발생하는 최소시간 : 3년
    + 통계자료에 의하면 주식 투자 기간이 3년이상이면 손실 가능성이 20% 정도로 드라마틱하게 줄고, 10년 이상 투자하면 손실가능성이 10% 미만으로 줄어든다.
    + 장기 투자해서 수익을 내기로 했다면, 목표 수익률을 얼마로 잡을 것인가? -> 09 ~ 17 년도 동안 연평균 8.8% 상승했었음.
    
    
- 삼성전자와 서브프라임 금융위기
    + 1998년 4월 27일 65,300원, 액면분할 직전인 2018년 4월 27일 2,650,000원이 되기까지 20년동안 주가가 무려 3,958% 만큼 상승했다.
    + 하지만 삼성전자 주식으로 부자가 되었다는 사람은 좀처럼 만나기 힘들다. 왜 그럴까?
    + 2008년 서브프라임 금융 위기때처럼 KOSPI 지수가 60% 이상 내려가는 극심한 하락장이 오면, 최고의 우량주에 투자했던 사람도 더 견디지 못하고 반토막 난 주식을 손해보고 판다. 그리고 시장을 떠난다.
    

- 백테스트 손실 예상
    + 투자하기 전에 기존 역사적 데이터를 이용하여 백테스트를 해 최대한 잃을 수 있는 손실을 미리 가늠해보고, 어느 만틈의 손실이 나면 매도헐지 손절매 기준을 확실히 정해야한다.
    + `손익 비대칭성의 원리`는 원금회복의 어려움을 표현한다.
    
- 헤지 전략
    + 투자할 때 발생하는 리스크나 손실을 줄이는 수단 또는 행위를 의미한다.
    + 자산별 투자시 자산간의 상관계수를 계산하여, 내가 투자하고자하는 종목과 -1에 가까운 자산을 함께 투자하는 것이 바람직하다.

- 직장인의 투자 규모
    + 회사 생활에 영향을 끼치는 것은 투자금액의 크기가 아니라 시장의 변동성을 감내할 수 있는 자신만의 투자 원칙이 있느냐에 달려있다.
    

- 퀀트 투자를 위한 파이썬
    + 퀀트 투자란 금융공학자들이 고도의 수학적, 통계적 지식을 바탕으로 파생상품의 가격을 모델링하여 투자하는 것을 의미했지만, 일반인들도 소프트웨어의 힘을 빌려 계량화된 수치에 따라 퀀트 투자를 할 수 있게 되었다.
    + 퀀트 투자의 장점 : 정량적 분석만으로 투자하기 때문에, 주관적인 요소가 개입될 여지가 적다.
    + 백테스트를 통해 내가 세운 투자 전략을 검증한다는 점이다.
